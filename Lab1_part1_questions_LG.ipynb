Для начала, давайте определимся с размерностью переменных. Пусть x - вектор размерности N, тогда y будет скалярной величиной.

Теперь, чтобы вычислить производную y по x, нам нужно применить правила дифференцирования скалярного произведения и матричного умножения.

y = x^T * x где x^T - транспонированная матрица x

Давайте развернем это выражение, чтобы упростить вычисления:

y = (x^T * x) = (x^T * x)^T = x^T * (x^T)^T = x^T * x

Мы получили, что y = x^T * x является симметричной матрицей.

Теперь давайте найдем производную этой функции по переменным x.

Для этого, мы можем использовать правило дифференцирования скалярного произведения:

d(y) / d(x) = d(x^T * x) / d(x) = (d(x^T) / d(x)) * x + x^T * (d(x) / d(x)) = (d(x^T) / d(x)) * x + x^T * I = x + x^T

Где I - единичная матрица размерности N.

Таким образом, производная функции y = x^T * x по переменным x равна x + x^T.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_derivative(x):
    return x + x.T

# Пример использования
N = 3
x = np.random.rand(N)
result = calculate_derivative(x)
print(result)

def calculate_derivative(x):
    return x + x.T

# Создание вектора x
N = 100
x = np.linspace(-10, 10, N)

# Вычисление производной
derivative = calculate_derivative(x)

# Построение графика
plt.plot(x, derivative)
plt.xlabel('x')
plt.ylabel('Derivative')
plt.title('Graph of the derivative of y = x^T * x')
plt.grid(True)
plt.show()
```

## ex. 2

$$ y = tr(AB) \quad A,B \in \mathbb{R}^{N \times N} $$

$$

нам нужно вычислить производную функции y = tr(AB) по матрице A.

Пусть A и B - матрицы размерности N x N, y - скалярная величина.

Для начала, давайте определимся с тем, что tr(AB) представляет собой след матричного произведения AB.

tr(AB) = ∑(i=1 to N) (AB)ii
где (AB)ii - элемент матрицы AB на пересечении i-ой строки и i-го столбца.

Теперь, чтобы вычислить производную y по матрице A, мы можем использовать правило дифференцирования следа матрицы.

Правило гласит:
d(tr(AB)) / dA = B^T
где B^T - транспонированная матрица B.

Таким образом, производная функции y = tr(AB) по матрице A равна B^T.

В этом примере, мы сначала импортируем библиотеку NumPy, которая предоставляет функциональность для работы с матрицами и векторами.

Затем, мы определяем функцию calculate_derivative, которая принимает матрицы A и B и вычисляет производную функции y = tr(AB) по матрице A.

Далее, мы создаем случайные матрицы A и B размерности N x N, и используем функцию calculate_derivative для вычисления производной.


```python

import numpy as np
import matplotlib.pyplot as plt

def calculate_derivative(A, B):
    return np.transpose(B)

# Создание матриц A и B
N = 100
A = np.random.rand(N, N)
B = np.random.rand(N, N)

# Вычисление производной
derivative = calculate_derivative(A, B)

# Построение графика
plt.imshow(derivative, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.xlabel('Column index')
plt.ylabel('Row index')
plt.title('Graph of the derivative of y = tr(AB) with respect to matrix A')
plt.show()
```

Функция y определяется как y = x^T * A * c, где x^T - это транспонирование вектора x.

Чтобы найти производную dy/dx, нам нужно продифференцировать y по x. Поскольку y является скалярным значением, а x - вектором, производная будет вектором той же размерности, что и x.

dy/dx = A * c

Чтобы найти производную dy/dA, нам нужно продифференцировать y по A. Поскольку y является скалярным значением, а A - матрицей, производная будет матрицей той же размерности, что и A.

dy/dA = x * c^T

Чтобы доказать подсказку, которая утверждает, что tr(ABC) = tr(CAB), мы можем использовать свойство циклической перестановки оператора следа. Свойство циклической перестановки утверждает, что tr(ABC) = tr(CAB) = tr(BCA).

Давайте реализуем эти уравнения на языке Python:



```python

import numpy as np

import numpy as np
import matplotlib.pyplot as plt

def calculate_derivative_x(A, c):
    return np.dot(A, c)

def calculate_derivative_A(x, c):
    return np.outer(x, c)

# Пример использования
N = 100
x = np.random.rand(N)
A = np.random.rand(N, N)
c = np.random.rand(N)

# Вычисление производных
derivative_x = calculate_derivative_x(A, c)
derivative_A = calculate_derivative_A(x, c)

# Построение графиков
plt.subplot(2, 1, 1)
plt.plot(x, derivative_x)
plt.xlabel('x')
plt.ylabel('dy/dx')
plt.title('График производной dy/dx')

plt.subplot(2, 1, 2)
plt.imshow(derivative_A, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.xlabel('Столбцы матрицы A')
plt.ylabel('Строки матрицы A')
plt.title('График производной dy/dA')

plt.tight_layout()
plt.show()
```
